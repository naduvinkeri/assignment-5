```python
import pandas as pd
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn import preprocessing
import numpy as np
from sklearn.linear_model import LinearRegression, Lasso
import datetime
from statistics import mean
```

```python
data= pd.read_csv('weatherHistory.csv')
```

```python
data["Date"]=pd.to_datetime(data["Formatted Date"],utc=True)
data["Month"]=data["Date"].dt.month
data
```

```python
data = data.drop(["Formatted Date","Daily Summary","Pressure (millibars)","Wind Speed (km/h)","Humidity","Loud Cover","Date"],axis=1)
data
```

```python
label_encoder = preprocessing.LabelEncoder()
data["Summary"]=label_encoder.fit_transform(data["Summary"])
data["Summary"].unique()
data
```

```python
data["Precip Type"]=data["Precip Type"].replace(["rain","snow"],[0,1])
data
```

```python
data.isna().sum()
```

```python
data["Precip Type"]=data["Precip Type"].replace(np.nan,0)
data.isna().sum()
```

```python
y=data["Apparent Temperature (C)"]
data = data.drop("Apparent Temperature (C)",axis =1)
print(data)
```

```python
y
```

```python
X_train, X_test,y_train,y_test=train_test_split(data,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
#finding variance using np
Variance = np.var(pred)
Variance
```

```python
SSE = np.mean((np.mean(pred)-y)**2)
Bias = SSE-Variance
Bias
```
